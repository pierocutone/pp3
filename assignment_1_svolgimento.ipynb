From the image you can see there are two parameters which are being varied for each sub-image. First, the rows are changed by color channel, where the top is the red channel, the middle is the green channel, and the bottom is the blue channel. Wait, why don't the colors look more red, green, and blue, in that order? Because the change you to be making is the ratio, or intensity, or that channel, in relationship to the other channels. We're going to use three different intensities, 0.1 (reduce the channel a lot), 0.5 (reduce the channel in half), and 0.9 (reduce the channel only a little bit).

For instance, a pixel represented as (200, 100, 50) is a sort of burnt orange color. So the top row of changes would create three alternative pixels, varying the first channel (red). one at (20, 100, 50), one at (100, 100, 50), and one at (180, 100, 50). The next row would vary the second channel (blue), and would create pixels of color values (200, 10, 50), (200, 50, 50) and (200, 90, 50).

```python
import PIL
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageDraw
from PIL import ImageFont


def get_concat_v(im1, im2):
    dst = Image.new('RGB', (im1.width, im1.height + im2.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    return dst


def append_text_to_image(image, text_message, text_box_height=60, font="readonly/fanwood-webfont.ttf", fontsize=70):
    # read font
    fnt = ImageFont.truetype(font, fontsize)

    # create text box
    text = Image.new('RGB', (image.width, text_box_height), (0,0,0))

    # get a drawing context
    d = ImageDraw.Draw(text)

    # draw text
    d.text((0,0), text_message, font=fnt)

    # append image and text
    image_with_text = get_concat_v(image, text)

    return image_with_text


def change_channel_intensity(image, channel, intensity):
    # initialize intensities
    r, g, b, = 1, 1, 1
    
    # modify intensity based on channel choice
    if channel == "R":
        r = intensity
    elif channel == "G":
        g = intensity
    elif channel == "B":
        b = intensity

    # create mask
    intensity_mask = ( 
        r, 0, 0, 0,
        0, g, 0, 0,
        0, 0, b, 0,
    )
    
    # modify image
    out = image.convert('RGB', intensity_mask)
    
    return out
```

```python
# create list of modified images
with Image.open("readonly/msi_recruitment.gif").convert("RGBA") as image:
    
    images = []
    channels = ["R", "G", "B"]
    intensities = [0.1, 0.5, 0.9]

    for channel in channels:
        for intensity in intensities:
            caption = "channel {} intensity {}".format(channels.index(channel), intensity)
            image_with_text = append_text_to_image(image, caption)
            image_c = change_channel_intensity(image_with_text, channel, intensity)
            images.append(image_c)
```

```python
# create a contact sheet
first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
x=0
y=0

for img in images:
    # paste the current image into the contact sheet
    contact_sheet.paste(img, (x, y) )
    # update X. If it is going to be the width of the image, then we set it to 0
    # update Y to point to the next "line" of the contact sheet
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
